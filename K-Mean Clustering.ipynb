```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
make_blobs
```

```python
plt.rcParams['figure.figsize']=(16,9)

X,y=make_blobs(n_samples=800, n_features=3, centers=3)
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2])
```

```python
X
```

```python
#n_cluster change thr no. of centers
kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
labels=kmeans.predict(X)
```

```python
labels
```

```python
centroidarray=kmeans.cluster_centers_
```

```python
centroidarray
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(X[:,0], X[:,1], X[:,2], c='y')
ax.scatter(centroidarray[:,0],centroidarray[:,1],centroidarray[:,2],marker='*',c='r',s=1000)
```

```python
import pandas as pd
url="http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

names=['sepal-length','sepal-width','petal-length','petal-width','Class']

dataset=pd.read_csv(url,names=names)
df=dataset
```

```python
df.head()
```

```python
df.info()
```

```python
df['Class']=pd.Categorical(df['Class'])
df['Class']=df['Class'].cat.codes

data=df.values[:,0:4]
category=df.values[:, 4]
```

```python
print(df.info)
print(type(category))
```

```python
data
```

```python
k=3

n=data.shape[0]
c=data.shape[1]
```

```python
#genetate random
mean_d=np.mean(data,axis=0)
std_d=np.std(data,axis=0)
```

```python
print(mean_d)
print(std_d)
```

```python
print(np.random.randn(k,c))
```

```python
centers=np.random.randn(k,c)*std_d+mean_d
print(centers)
```

```python
colors=['orange','blue','green']
```

```python
for i in range(n):
    plt.scatter(data[i,0],data[i,1],s=10,color=colors[int(category[i])])
plt.scatter(centers[:,0],centers[:,1],marker='*',c='r',s=150)
```

```python
from copy import deepcopy 

centers_old=np.zeros(centers.shape)
centers_new= deepcopy(centers)
```

```python
print(centers_old)
print(centers_new)
print(data.shape)
```

```python
clusters=np.zeros(n)
distances=np.zeros((n,k))

print(clusters)
print(distances)
```

```python
error=np.linalg.norm(centers_new - centers_old)
error
```

```python
centers_new - centers_old
```

```python
while error !=0:
    #measure the distance to every centers
    for i in range(k):
        distances[:,i]=np.linalg.norm(data - centers[i],axis=1)
    # assign all traning to closet center
    clusters=np.argmin(distances,axis=1)
    
    centers_old=deepcopy(centers_new)
    # calculate mean for every cluster and update the center
    for i in range(k):
        centers_new[i]=np.mean(data[clusters==i],axis=0)
    error=np.linalg.norm(centers_new - centers_old)
    print(error)
    
print(centers_new)
```

```python
colors=['black','blue','red']

for i in range(n):
    plt.scatter(data[i,0], data[i,3], s=7, color=colors[int(category[i])])
plt.scatter(centers[:,0],centers[:,3], marker='*', c='g', s=150)
```
